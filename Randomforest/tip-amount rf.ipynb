```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("/content/drive/MyDrive/Dataset/tip-amount_(2).xlsx")
```

```python
data.head()
```

```python
corrmatrix = data.corr()
plt.subplots(figsize=(10,4))
sns.heatmap(corrmatrix,vmin=-2,vmax=1,annot=True,linewidths=0.2,cmap='YlGnBu')
```

```python
data.columns
```

```python
y=data['Observed tip bill(yi)']
x=data.drop(['Observed tip bill(yi)','Meal'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
model=rf.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
predictions
```

```python
print(y_test)
```

```python
print(predictions)
```

```python
from sklearn.metrics import mean_squared_error
print('MSC is :',mean_squared_error(y_test,predictions))
```
